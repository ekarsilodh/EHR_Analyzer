```python
import os, re, json, math, sys, subprocess
from pathlib import Path
from collections import Counter
import urllib.request

import numpy as np
import pandas as pd

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler

from sklearn.model_selection import GroupShuffleSplit, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    roc_auc_score, average_precision_score,
    precision_recall_fscore_support, precision_recall_curve,
    classification_report, confusion_matrix,
    roc_curve
)

import matplotlib.pyplot as plt


# AUTO-DOWNLOAD DATA FROM GITHUB

GITHUB_REPO = "abhijith-one77/ML_project"
GITHUB_BRANCH = "main"          # change to "master" if needed
DATA_DIR = Path("/mnt/data")    # where to store downloaded csvs
DATA_DIR.mkdir(parents=True, exist_ok=True)

FORCE_DOWNLOAD = False          # set True to re-download even if files exist

# Try these candidate locations in the repo (edit if your CSVs are elsewhere)
CANDIDATE_SUBDIRS = [
    "",          # repo root
    "data",
    "dataset",
    "datasets",
]

REQUIRED_FILES = [
    "patients_table.csv",
    "patient_admissions.csv",
    "disease_diagnosis_code.csv",
]

def _raw_url(repo: str, branch: str, subdir: str, filename: str) -> str:
    # https://raw.githubusercontent.com/<owner>/<repo>/<branch>/<path>
    if subdir:
        return f"https://raw.githubusercontent.com/{repo}/{branch}/{subdir}/{filename}"
    return f"https://raw.githubusercontent.com/{repo}/{branch}/{filename}"

def download_if_missing(local_path: Path, filename: str) -> None:
    if local_path.exists() and not FORCE_DOWNLOAD:
        print(f"[OK] Found {filename} -> {local_path}")
        return

    last_err = None
    for subdir in CANDIDATE_SUBDIRS:
        url = _raw_url(GITHUB_REPO, GITHUB_BRANCH, subdir, filename)
        try:
            print(f"[download] {filename} <- {url}")
            urllib.request.urlretrieve(url, local_path.as_posix())
            if local_path.stat().st_size < 10:
                raise RuntimeError("Downloaded file is unexpectedly small.")
            print(f"[OK] Saved {filename} -> {local_path}")
            return
        except Exception as e:
            last_err = e

    raise RuntimeError(
        f"Could not download '{filename}' from repo '{GITHUB_REPO}' (branch '{GITHUB_BRANCH}').\n"
        f"Tried subdirs: {CANDIDATE_SUBDIRS}\n"
        f"Last error: {repr(last_err)}\n"
        f"Fix: set GITHUB_BRANCH correctly and/or update CANDIDATE_SUBDIRS to the folder containing the CSVs."
    )

# Ensure required CSVs exist locally
for fn in REQUIRED_FILES:
    download_if_missing(DATA_DIR / fn, fn)

# -------------------------
# CONFIG
# -------------------------
SEED = 42
MAX_ADM = 20          # admissions per patient (time steps)
MAX_DX  = 40          # diagnosis tokens per admission
MIN_FREQ = 5          # vocab pruning threshold (TRAIN ONLY)
DROPOUT = 0.15

# Use the downloaded paths
PATIENTS_PATH   = str(DATA_DIR / "patients_table.csv")
ADMISSIONS_PATH = str(DATA_DIR / "patient_admissions.csv")
DX_PATH         = str(DATA_DIR / "disease_diagnosis_code.csv")

OUT_NPZ    = str(DATA_DIR / "seq_feats_lstm_transformer.npz")
VOCAB_JSON = str(DATA_DIR / "dx_vocab.json")
ENC_JSON   = str(DATA_DIR / "cat_encoders.json")
PLOT_DIR   = str(DATA_DIR / "plots")

# training
BATCH_SIZE = 32
VAL_BATCH_SIZE = 64
TEST_BATCH_SIZE = 64

# multitask
ALPHA_LOS = 0.05

LR_LSTM  = 8e-4
LR_TRANS = 6e-4
WD_LSTM  = 1e-5
WD_TRANS = 5e-4

PATIENCE = 10
MIN_DELTA = 1e-4
EPOCHS = 60

# Imbalance handling (choose ONE strategy)
USE_WEIGHTED_SAMPLER = True
USE_POS_WEIGHT_BCE   = False

# Threshold selection
MIN_PRED_POS_VAL = 10

# ICD mapping: best-effort ICD9 -> ICD10 3-char category
USE_ICD_MAPPINGS = True
UNK_DX = "UNK_DX"

np.random.seed(SEED)
torch.manual_seed(SEED)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)


# -------------------------
# HELPERS
# -------------------------
def drop_unnamed(df: pd.DataFrame) -> pd.DataFrame:
    return df.loc[:, ~df.columns.str.contains(r"^Unnamed")].copy()

def to_dt(s):
    return pd.to_datetime(s, errors="coerce")

def ensure_int(df, col):
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce").fillna(0).astype(np.int64)
    return df

def first_existing(df, candidates, default=None):
    for c in candidates:
        if c in df.columns:
            return c
    return default

def factorize(series: pd.Series):
    s = series.fillna("UNKNOWN").astype(str)
    cats = sorted(s.unique().tolist())
    mapping = {c: i for i, c in enumerate(cats)}
    codes = s.map(mapping).astype(int)
    return codes, mapping

def safe_get_col(df, col, default_value="UNKNOWN"):
    if col and col in df.columns:
        return df[col]
    return pd.Series([default_value] * len(df), index=df.index)

def clean_icd(code):
    if pd.isna(code):
        return ""
    s = str(code).strip().upper()
    s = s.replace(".", "")
    s = re.sub(r"[^A-Z0-9]", "", s)
    return s

# --- ICD9 -> ICD10 mapping (best-effort) ---
def ensure_icd_mappings_installed():
    """
    Try to import icdmappings; if missing, attempt pip install.
    If install/import fails (e.g., no internet), return False.
    """
    try:
        from icdmappings import Mapper  # noqa
        return True
    except Exception:
        try:
            subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "icd-mappings"])
            from icdmappings import Mapper  # noqa
            return True
        except Exception as e:
            print("WARNING: Could not install/import icd-mappings. Falling back to heuristic ICD handling.")
            print("  Error:", repr(e))
            return False

def icd10_3char_from_icd10(icd_norm: str):
    if not icd_norm:
        return None
    m = re.match(r"^([A-Z]\d{2})", icd_norm)
    return m.group(1) if m else (icd_norm[:3] if len(icd_norm) >= 3 else None)

def map_icd9_to_icd10_3char(mapper, icd9_norm: str):
    if not icd9_norm:
        return None
    try:
        mapped = mapper.map([icd9_norm], source="icd9", target="icd10")
        # mapped may be dict-like or list-like depending on version
        if isinstance(mapped, dict):
            cand = mapped.get(icd9_norm, None)
        else:
            cand = mapped

        if isinstance(cand, list) and len(cand) > 0:
            for c in cand:
                c_norm = clean_icd(c)
                tok = icd10_3char_from_icd10(c_norm)
                if tok:
                    return tok
            return None

        if isinstance(cand, str):
            c_norm = clean_icd(cand)
            return icd10_3char_from_icd10(c_norm)

        return None
    except Exception:
        return None

def unified_dx_token_icd10only(icd_norm: str, icd_version, mapper=None):
    """
    Returns a single token space in ICD-10 3-char categories.
    - ICD-10: use ICD-10 3-char
    - ICD-9 : map to ICD-10 then 3-char (best-effort)
    - If mapping fails: UNK_DX
    """
    if not icd_norm:
        return None
    v = str(icd_version)

    # Treat ICD-10 or "ICD-10-like" strings as ICD-10
    if v == "10" or (icd_norm[:1].isalpha() and re.match(r"^[A-Z]\d", icd_norm)):
        tok = icd10_3char_from_icd10(icd_norm)
        return tok if tok else UNK_DX

    # ICD-9
    if mapper is not None:
        tok = map_icd9_to_icd10_3char(mapper, icd_norm)
        return tok if tok else UNK_DX

    # fallback without mapper
    return UNK_DX


# -------------------------
# PLOT HELPERS (save + combined ROC)
# -------------------------
def ensure_plot_dir():
    Path(PLOT_DIR).mkdir(parents=True, exist_ok=True)

def savefig(path):
    Path(os.path.dirname(path)).mkdir(parents=True, exist_ok=True)
    plt.savefig(path, dpi=200, bbox_inches="tight")

def roc_points(y_true, y_prob):
    fpr, tpr, _ = roc_curve(y_true, y_prob)
    aucv = roc_auc_score(y_true, y_prob) if (np.min(y_true) != np.max(y_true)) else np.nan
    return fpr, tpr, aucv

def plot_and_save_roc(y_true, y_prob, title, outpath, label=None):
    fpr, tpr, aucv = roc_points(y_true, y_prob)
    plt.figure(figsize=(6, 5))
    if label is None:
        label = f"AUROC={aucv:.4f}"
    plt.plot(fpr, tpr, label=label)
    plt.plot([0, 1], [0, 1], linestyle="--")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(title)
    plt.legend()
    plt.tight_layout()
    savefig(outpath)
    plt.show()
    return fpr, tpr, aucv

def plot_and_save_confusion(y_true, y_pred, title, outpath):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(5.5, 4.5))
    plt.imshow(cm, interpolation="nearest")
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(2)
    plt.xticks(tick_marks, ["0", "1"])
    plt.yticks(tick_marks, ["0", "1"])

    thresh = cm.max() / 2.0 if cm.max() > 0 else 0
    for i in range(2):
        for j in range(2):
            plt.text(j, i, format(cm[i, j], "d"),
                     ha="center", va="center",
                     color="white" if cm[i, j] > thresh else "black")

    plt.ylabel("True label")
    plt.xlabel("Predicted label")
    plt.tight_layout()
    savefig(outpath)
    plt.show()

def sigmoid_np(x):
    return 1.0 / (1.0 + np.exp(-x))


# -------------------------
# 1) LOAD
# -------------------------
patients   = drop_unnamed(pd.read_csv(PATIENTS_PATH))
admissions = drop_unnamed(pd.read_csv(ADMISSIONS_PATH))
dx         = drop_unnamed(pd.read_csv(DX_PATH))

patients   = ensure_int(patients, "subject_id")
admissions = ensure_int(admissions, "subject_id")
admissions = ensure_int(admissions, "hadm_id")
dx         = ensure_int(dx, "subject_id")
dx         = ensure_int(dx, "hadm_id")

# Datetimes
admittime_col = first_existing(admissions, ["admittime", "admit_time", "admission_time"])
dischtime_col = first_existing(admissions, ["dischtime", "disch_time", "discharge_time"])
deathtime_col = first_existing(admissions, ["deathtime", "death_time"])

admissions["admittime_dt"] = to_dt(admissions[admittime_col]) if admittime_col else pd.NaT
admissions["dischtime_dt"] = to_dt(admissions[dischtime_col]) if dischtime_col else pd.NaT
admissions["deathtime_dt"] = to_dt(admissions[deathtime_col]) if deathtime_col else pd.NaT


# -------------------------
# 2) LABELS (do NOT leak these into features)
# -------------------------
mort_col = first_existing(admissions, ["hospital_expire_flag", "mortality", "death", "died"])
if mort_col is None:
    raise ValueError("Admissions missing mortality target (expected hospital_expire_flag or similar).")
admissions["mortality"] = pd.to_numeric(admissions[mort_col], errors="coerce").fillna(0).astype(int)

# LOS label (hours) from times
los_from_times = (admissions["dischtime_dt"] - admissions["admittime_dt"]).dt.total_seconds() / 3600.0
if "los_hours" in admissions.columns:
    los_existing = pd.to_numeric(admissions["los_hours"], errors="coerce")
    admissions["los_hours"] = los_existing.where(los_existing.notna(), los_from_times)
else:
    admissions["los_hours"] = los_from_times

admissions["los_hours"] = pd.to_numeric(admissions["los_hours"], errors="coerce").clip(lower=0).fillna(0.0)
admissions["los_log"] = np.log1p(admissions["los_hours"])


# -------------------------
# 3) SAFE TIME FEATURES (admission-time only)
# -------------------------
admissions["adm_year"]    = admissions["admittime_dt"].dt.year.fillna(0).astype(int)
admissions["adm_month"]   = admissions["admittime_dt"].dt.month.fillna(0).astype(int)
admissions["adm_day"]     = admissions["admittime_dt"].dt.day.fillna(0).astype(int)
admissions["adm_hour"]    = admissions["admittime_dt"].dt.hour.fillna(0).astype(int)
admissions["adm_weekday"] = admissions["admittime_dt"].dt.weekday.fillna(0).astype(int)
admissions["adm_weekend"] = (admissions["adm_weekday"] >= 5).astype(int)


# -------------------------
# 4) MERGE PATIENT DEMOGRAPHICS
# -------------------------
gender_col_p = first_existing(patients, ["gender", "sex"])
age_col_p    = first_existing(patients, ["anchor_age", "age"])
year_col_p   = first_existing(patients, ["anchor_year"])
dod_col_p    = first_existing(patients, ["dod", "date_of_death"])

keep_p_cols = ["subject_id"]
for c in [gender_col_p, age_col_p, year_col_p, dod_col_p]:
    if c and c not in keep_p_cols:
        keep_p_cols.append(c)

p_small = patients[keep_p_cols].copy()
df = admissions.merge(p_small, on="subject_id", how="left")

# normalize expected columns
if gender_col_p and gender_col_p != "gender":
    df["gender"] = df[gender_col_p]
elif "gender" not in df.columns:
    df["gender"] = "UNKNOWN"

if age_col_p and age_col_p != "anchor_age":
    df["anchor_age"] = df[age_col_p]
elif "anchor_age" not in df.columns:
    df["anchor_age"] = 0

df["age_at_admit"] = pd.to_numeric(df["anchor_age"], errors="coerce").fillna(0.0)

race_col = first_existing(df, ["race", "ethnicity"])
if race_col and race_col != "race":
    df["race"] = df[race_col].fillna("UNKNOWN").astype(str)
elif "race" not in df.columns:
    df["race"] = "UNKNOWN"

admtype_col   = first_existing(df, ["admission_type", "adm_type"])
insurance_col = first_existing(df, ["insurance"])
marital_col   = first_existing(df, ["marital_status", "marital"])


# -------------------------
# 5) DIAGNOSIS TOKENS PER ADMISSION (ICD9 -> ICD10 harmonised best-effort)
# -------------------------
if "icd_code" not in dx.columns:
    raise ValueError("DX file missing icd_code column.")
if "icd_version" not in dx.columns:
    dx["icd_version"] = 10

dx["icd_norm"] = dx["icd_code"].apply(clean_icd)

mapper = None
if USE_ICD_MAPPINGS:
    ok = ensure_icd_mappings_installed()
    if ok:
        from icdmappings import Mapper
        mapper = Mapper()

dx["dx_token"] = [
    unified_dx_token_icd10only(c, v, mapper=mapper)
    for c, v in zip(dx["icd_norm"], dx["icd_version"])
]
dx = dx.dropna(subset=["dx_token"]).copy()

dx["seq_num"] = pd.to_numeric(dx.get("seq_num", 0), errors="coerce").fillna(0).astype(int)
dx.sort_values(["subject_id", "hadm_id", "seq_num"], inplace=True)

dx_lists = dx.groupby(["subject_id", "hadm_id"])["dx_token"].apply(list).reset_index()

if "hadm_id" not in df.columns:
    raise ValueError("Admissions missing hadm_id column (required to attach diagnoses).")

df = df.merge(dx_lists, on=["subject_id", "hadm_id"], how="left")
df["dx_token"] = df["dx_token"].apply(lambda x: x if isinstance(x, list) else [])
df["n_dx"] = df["dx_token"].apply(len)
df["n_dx_unique"] = df["dx_token"].apply(lambda lst: len(set(lst)))


# -------------------------
# 6) PER-PATIENT SEQUENCE ORDER + SAFE DELTA-TIME FEATURE
# -------------------------
df["admittime_dt"] = pd.to_datetime(df["admittime_dt"], errors="coerce")
df["_admit_sort_key"] = df["admittime_dt"].fillna(pd.Timestamp("1900-01-01"))
df.sort_values(["subject_id", "_admit_sort_key"], inplace=True)
df.drop(columns=["_admit_sort_key"], inplace=True)

df["prev_admittime_dt"] = df.groupby("subject_id")["admittime_dt"].shift(1)
delta = (df["admittime_dt"] - df["prev_admittime_dt"]).dt.total_seconds() / 3600.0
df["delta_prev_adm_hours"] = pd.to_numeric(delta, errors="coerce").fillna(0.0).clip(lower=0.0)


# -------------------------
# 7) ENCODE CATEGORICALS
# -------------------------
df["gender_code"], gender_map = factorize(df["gender"])
df["race_code"], race_map     = factorize(df["race"])

df["admission_type_code"], admtype_map = factorize(safe_get_col(df, admtype_col, "UNKNOWN"))
df["insurance_code"], ins_map          = factorize(safe_get_col(df, insurance_col, "UNKNOWN"))
df["marital_code"], mar_map            = factorize(safe_get_col(df, marital_col, "UNKNOWN"))

encoders = {
    "gender": gender_map,
    "race": race_map,
    "admission_type": admtype_map,
    "insurance": ins_map,
    "marital_status": mar_map
}

# IMPORTANT: Do NOT include dischtime/deathtime/los as FEATURES (leakage).
static_cols = [
    "age_at_admit",
    "gender_code",
    "race_code",
    "admission_type_code",
    "insurance_code",
    "marital_code",
    "adm_year", "adm_month", "adm_day", "adm_hour", "adm_weekday", "adm_weekend",
    "n_dx", "n_dx_unique",
    "delta_prev_adm_hours",
]
for c in static_cols:
    df[c] = pd.to_numeric(df[c], errors="coerce").fillna(0.0)


# -------------------------
# 8) PATIENT-LEVEL SPLITS (no leakage)
# -------------------------
subjects_all = df["subject_id"].unique()

# patient-level label = last admission mortality
last_rows = df.groupby("subject_id", sort=False).tail(1)
y_patient = last_rows.set_index("subject_id")["mortality"].reindex(subjects_all).fillna(0).astype(int).values

gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=SEED)
train_idx, test_idx = next(gss.split(np.zeros((len(subjects_all), 1)), y_patient, groups=subjects_all))
subj_train_all = subjects_all[train_idx]
subj_test = subjects_all[test_idx]
subj_tr, subj_val = train_test_split(subj_train_all, test_size=0.2, random_state=SEED)

def patient_mask(subjs):
    return df["subject_id"].isin(set(subjs))

df_train = df.loc[patient_mask(subj_tr)].copy()
df_val   = df.loc[patient_mask(subj_val)].copy()
df_test  = df.loc[patient_mask(subj_test)].copy()

print("TRAIN label distribution:", Counter(df_train.groupby("subject_id").tail(1)["mortality"].astype(int).tolist()))
print("VAL   label distribution:", Counter(df_val.groupby("subject_id").tail(1)["mortality"].astype(int).tolist()))
print("TEST  label distribution:", Counter(df_test.groupby("subject_id").tail(1)["mortality"].astype(int).tolist()))


# -------------------------
# 9) TRAIN-ONLY VOCAB
# -------------------------
PAD = "<PAD>"
UNK = "<UNK>"

freq = Counter()
for lst in df_train["dx_token"]:
    freq.update(lst)

vocab = {PAD: 0, UNK: 1}
for tok, ct in freq.items():
    if ct >= MIN_FREQ:
        vocab[tok] = len(vocab)

def tokens_to_ids(tokens):
    ids = [vocab.get(t, vocab[UNK]) for t in tokens[:MAX_DX]]
    return ids

df["dx_ids"] = df["dx_token"].apply(tokens_to_ids)


# -------------------------
# 10) BUILD PER-PATIENT TENSORS
# -------------------------
subjects = subjects_all.tolist()
N = len(subjects)
D_static = len(static_cols)

dx_ids   = np.zeros((N, MAX_ADM, MAX_DX), dtype=np.int32)
dx_mask  = np.zeros((N, MAX_ADM, MAX_DX), dtype=np.int8)
adm_stat = np.zeros((N, MAX_ADM, D_static), dtype=np.float32)
adm_mask = np.zeros((N, MAX_ADM), dtype=np.int8)

y_mort   = np.zeros((N,), dtype=np.int64)
y_loslog = np.zeros((N,), dtype=np.float32)
race_last = np.array(["UNKNOWN"] * N, dtype=object)

grouped = df.groupby("subject_id", sort=False)

for i, sid in enumerate(subjects):
    g = grouped.get_group(sid).tail(MAX_ADM)
    if len(g) == 0:
        continue
    last = g.iloc[-1]
    y_mort[i]   = int(last["mortality"])
    y_loslog[i] = float(last["los_log"])
    race_last[i]= str(last["race"])

    for t, (_, row) in enumerate(g.iterrows()):
        adm_mask[i, t] = 1
        adm_stat[i, t, :] = row[static_cols].values.astype(np.float32)

        ids = row["dx_ids"]
        L = min(len(ids), MAX_DX)
        if L > 0:
            dx_ids[i, t, :L] = np.asarray(ids[:L], dtype=np.int32)
            dx_mask[i, t, :L] = 1

Path(str(DATA_DIR)).mkdir(parents=True, exist_ok=True)
np.savez_compressed(
    OUT_NPZ,
    dx_ids=dx_ids,
    dx_mask=dx_mask,
    adm_static=adm_stat,
    adm_mask=adm_mask,
    y_mort=y_mort,
    y_los_log=y_loslog,
    race_last=race_last.astype(str),
    subjects=np.asarray(subjects, dtype=np.int64),
    static_cols=np.asarray(static_cols, dtype=object),
)
Path(VOCAB_JSON).write_text(json.dumps(vocab, indent=2))
Path(ENC_JSON).write_text(json.dumps(encoders, indent=2))

print("\nSaved:", OUT_NPZ)
print("Vocab size:", len(vocab))
print("Shapes:", dx_ids.shape, dx_mask.shape, adm_stat.shape, adm_mask.shape)

VOCAB_SIZE = max(int(dx_ids.max()) + 1, 2)


# -------------------------
# 11) SPLIT INDICES (patient-level)
# -------------------------
subjects_np = np.asarray(subjects, dtype=np.int64)

idx_train = np.where(np.isin(subjects_np, subj_tr.astype(np.int64)))[0]
idx_val   = np.where(np.isin(subjects_np, subj_val.astype(np.int64)))[0]
idx_test  = np.where(np.isin(subjects_np, subj_test.astype(np.int64)))[0]


# -------------------------
# 12) SCALE STATIC FEATURES (fit on TRAIN admissions only)
# -------------------------
train_static_flat = adm_stat[idx_train].reshape(-1, D_static)
train_adm_mask_flat = adm_mask[idx_train].reshape(-1)
real = (train_adm_mask_flat == 1)

scaler = StandardScaler()
scaler.fit(train_static_flat[real] if real.any() else train_static_flat)

adm_static_scaled = np.zeros_like(adm_stat, dtype=np.float32)
for i in range(N):
    x = adm_stat[i].copy()
    m = adm_mask[i].astype(bool)
    if m.any():
        x[m] = scaler.transform(x[m])
    x[~m] = 0.0
    adm_static_scaled[i] = x


# -------------------------
# 13) DATASET + SAMPLER
# -------------------------
class SeqEHRDataset(Dataset):
    def __init__(self, idxs):
        self.idxs = np.array(idxs, dtype=int)

    def __len__(self):
        return len(self.idxs)

    def __getitem__(self, k):
        i = self.idxs[k]
        return (
            torch.tensor(dx_ids[i], dtype=torch.long),                 # [T,L]
            torch.tensor(dx_mask[i], dtype=torch.float32),             # [T,L]
            torch.tensor(adm_static_scaled[i], dtype=torch.float32),   # [T,D]
            torch.tensor(adm_mask[i], dtype=torch.float32),            # [T]
            torch.tensor([float(y_mort[i])], dtype=torch.float32),     # [1]
            torch.tensor([float(y_loslog[i])], dtype=torch.float32),   # [1]
            str(race_last[i]),
        )

def collate_fixed(batch):
    x_dx, x_dxm, x_stat, x_am, y_m, y_l, race = zip(*batch)
    return (
        torch.stack(x_dx),
        torch.stack(x_dxm),
        torch.stack(x_stat),
        torch.stack(x_am),
        torch.stack(y_m),
        torch.stack(y_l),
        list(race),
    )

train_ds = SeqEHRDataset(idx_train)
val_ds   = SeqEHRDataset(idx_val)
test_ds  = SeqEHRDataset(idx_test)

sampler = None
if USE_WEIGHTED_SAMPLER:
    ytr = y_mort[idx_train].astype(int)
    c0 = (ytr == 0).sum()
    c1 = (ytr == 1).sum()
    w0 = 1.0 / max(1, c0)
    w1 = 1.0 / max(1, c1)
    weights = np.where(ytr == 1, w1, w0).astype(np.float64)
    sampler = WeightedRandomSampler(weights=weights, num_samples=len(weights), replacement=True)
    shuffle = False
else:
    shuffle = True

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE,
                          sampler=sampler, shuffle=shuffle, collate_fn=collate_fixed)
val_loader   = DataLoader(val_ds, batch_size=VAL_BATCH_SIZE,
                          shuffle=False, collate_fn=collate_fixed)
test_loader  = DataLoader(test_ds, batch_size=TEST_BATCH_SIZE,
                          shuffle=False, collate_fn=collate_fixed)


# -------------------------
# 14) ADMISSION ENCODER (attention pooling over dx tokens)
# -------------------------
class DxAttentionPool(nn.Module):
    """Token-level attention pooling within each admission."""
    def __init__(self, emb_dim, dropout=DROPOUT):
        super().__init__()
        self.attn = nn.Sequential(
            nn.Linear(emb_dim, emb_dim),
            nn.Tanh(),
            nn.Linear(emb_dim, 1)
        )
        self.drop = nn.Dropout(dropout)

    def forward(self, emb, mask):
        # emb: [B,T,L,E], mask: [B,T,L] {0,1}
        logits = self.attn(self.drop(emb)).squeeze(-1)  # [B,T,L]
        logits = logits.masked_fill(mask <= 0, -1e9)
        w = torch.softmax(logits, dim=-1).unsqueeze(-1)  # [B,T,L,1]
        pooled = (emb * w).sum(dim=2)  # [B,T,E]
        return pooled

class AdmissionEncoder(nn.Module):
    def __init__(self, vocab_size, dx_emb_dim=64, static_dim=None, dropout=0.15):
        super().__init__()
        if static_dim is None:
            raise ValueError("AdmissionEncoder needs static_dim set.")
        self.dx_emb = nn.Embedding(vocab_size, dx_emb_dim, padding_idx=0)
        self.pool = DxAttentionPool(dx_emb_dim, dropout=dropout)
        self.norm = nn.LayerNorm(dx_emb_dim + static_dim)
        self.drop = nn.Dropout(dropout)
        self.out_dim = dx_emb_dim + static_dim

    def forward(self, dx_ids, dx_mask, static, adm_mask):
        e = self.dx_emb(dx_ids)        # [B,T,L,E]
        pooled = self.pool(e, dx_mask) # [B,T,E]
        x = torch.cat([pooled, static], dim=-1)   # [B,T,E+D]
        x = x * adm_mask.unsqueeze(-1)            # zero padded admissions
        x = self.norm(x)
        x = self.drop(x)
        return x


# -------------------------
# 15) MODELS
# -------------------------
class ClinicalMultiTaskLSTM(nn.Module):
    def __init__(self, input_dim, hidden_dim=128):
        super().__init__()
        self.lstm = nn.LSTM(
            input_dim, hidden_dim, num_layers=2,
            batch_first=True, bidirectional=True, dropout=0.25
        )
        self.mort_head = nn.Sequential(nn.Linear(hidden_dim * 2, 64), nn.ReLU(), nn.Dropout(0.2), nn.Linear(64, 1))
        self.los_head  = nn.Sequential(nn.Linear(hidden_dim * 2, 64), nn.ReLU(), nn.Dropout(0.2), nn.Linear(64, 1))

    def forward(self, x, lengths):
        packed = nn.utils.rnn.pack_padded_sequence(x, lengths.cpu(), batch_first=True, enforce_sorted=False)
        _, (h_n, _) = self.lstm(packed)
        feat = torch.cat((h_n[-2, :, :], h_n[-1, :, :]), dim=1)
        return self.mort_head(feat), self.los_head(feat)

class PositionalEncoding(nn.Module):
    def __init__(self, d_model, max_len=MAX_ADM + 1):
        super().__init__()
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        self.register_buffer("pe", pe.unsqueeze(0))

    def forward(self, x):
        return x + self.pe[:, :x.size(1)]

class ClinicalTransformer(nn.Module):
    def __init__(self, input_dim, embed_dim=128, nhead=4, num_layers=2, dropout=0.25):
        super().__init__()
        self.embed_dim = embed_dim
        self.proj = nn.Linear(input_dim, embed_dim)
        self.cls_token = nn.Parameter(torch.zeros(1, 1, embed_dim))
        self.pos = PositionalEncoding(embed_dim, max_len=MAX_ADM + 1)

        enc_layer = nn.TransformerEncoderLayer(
            d_model=embed_dim, nhead=nhead, dim_feedforward=embed_dim * 2,
            dropout=dropout, batch_first=True, norm_first=False
        )
        self.encoder = nn.TransformerEncoder(enc_layer, num_layers=num_layers)

        self.mort_head = nn.Sequential(nn.Linear(embed_dim, 64), nn.ReLU(), nn.Dropout(dropout), nn.Linear(64, 1))
        self.los_head  = nn.Sequential(nn.Linear(embed_dim, 64), nn.ReLU(), nn.Dropout(dropout), nn.Linear(64, 1))

    def forward(self, x, adm_mask):
        pad_mask = (adm_mask == 0)  # [B,T]
        x = self.proj(x)            # [B,T,E]
        B = x.size(0)
        cls = self.cls_token.expand(B, 1, self.embed_dim)
        x = torch.cat([cls, x], dim=1)  # [B,T+1,E]

        cls_pad = torch.zeros((B, 1), dtype=torch.bool, device=x.device)
        src_key_padding_mask = torch.cat([cls_pad, pad_mask.to(x.device)], dim=1)  # [B,T+1]

        x = self.pos(x)
        out = self.encoder(x, src_key_padding_mask=src_key_padding_mask)
        feat = out[:, 0, :]
        return self.mort_head(feat), self.los_head(feat)

class LSTMWrapper(nn.Module):
    def __init__(self, adm_encoder, lstm_model):
        super().__init__()
        self.adm_encoder = adm_encoder
        self.model = lstm_model
    def forward(self, dx_ids, dx_mask, static, adm_mask):
        x = self.adm_encoder(dx_ids, dx_mask, static, adm_mask)
        lengths = adm_mask.sum(dim=1).clamp(min=1).long()
        return self.model(x, lengths)

class TransWrapper(nn.Module):
    def __init__(self, adm_encoder, trans_model):
        super().__init__()
        self.adm_encoder = adm_encoder
        self.model = trans_model
    def forward(self, dx_ids, dx_mask, static, adm_mask):
        x = self.adm_encoder(dx_ids, dx_mask, static, adm_mask)
        return self.model(x, adm_mask)


# -------------------------
# 16) LOSSES
# -------------------------
if USE_POS_WEIGHT_BCE:
    pos_count = int(y_mort[idx_train].sum())
    pos_w_val = len(idx_train) / max(1, pos_count)
    pos_w = torch.tensor([pos_w_val], dtype=torch.float32).to(device)
else:
    pos_w = None

bce = nn.BCEWithLogitsLoss(pos_weight=pos_w) if pos_w is not None else nn.BCEWithLogitsLoss()
los_loss_fn = nn.SmoothL1Loss(beta=0.5)


# -------------------------
# 17) EARLY STOPPING (VAL AUPRC)
# -------------------------
class EarlyStoppingAUPRC:
    def __init__(self, patience=PATIENCE, min_delta=MIN_DELTA):
        self.patience = patience
        self.min_delta = min_delta
        self.best = -1.0
        self.counter = 0
        self.stop = False

    def step(self, metric):
        if np.isnan(metric):
            metric = -1.0
        if metric > self.best + self.min_delta:
            self.best = metric
            self.counter = 0
        else:
            self.counter += 1
            if self.counter >= self.patience:
                self.stop = True
        return self.stop


# -------------------------
# 18) INFERENCE + THRESHOLD + SUBGROUPS
# -------------------------
def collect_logits_probs_and_los(model, loader):
    model.eval()
    all_logits, all_prob, all_y = [], [], []
    all_lhat_hours, all_ltrue_hours = [], []
    all_races = []
    with torch.no_grad():
        for dx_b, dxm_b, stat_b, am_b, ymb, ylb, races in loader:
            dx_b  = dx_b.to(device)
            dxm_b = dxm_b.to(device)
            stat_b = stat_b.to(device)
            am_b  = am_b.to(device)

            pm, pl = model(dx_b, dxm_b, stat_b, am_b)

            logits = pm.detach().cpu().numpy().ravel()
            prob = sigmoid_np(logits)

            pred_hours = np.expm1(pl.detach().cpu().numpy().ravel())
            true_hours = np.expm1(ylb.numpy().ravel())

            all_logits.extend(list(logits))
            all_prob.extend(list(prob))
            all_y.extend(list(ymb.numpy().ravel().astype(int)))
            all_lhat_hours.extend(list(pred_hours))
            all_ltrue_hours.extend(list(true_hours))
            all_races.extend(list(races))

    return (np.array(all_logits), np.array(all_prob), np.array(all_y).astype(int),
            np.array(all_lhat_hours), np.array(all_ltrue_hours), np.array(all_races))

def pick_threshold_max_f1_min_pos(y_true, y_prob, min_pos=MIN_PRED_POS_VAL):
    y_true = np.asarray(y_true).astype(int)
    y_prob = np.asarray(y_prob).astype(float)
    prec, rec, thr = precision_recall_curve(y_true, y_prob)

    best_t, best_f1 = 0.5, -1.0
    for t in thr:
        yhat = (y_prob >= t).astype(int)
        if yhat.sum() < min_pos:
            continue
        p, r, f1, _ = precision_recall_fscore_support(y_true, yhat, average="binary", zero_division=0)
        if f1 > best_f1:
            best_f1, best_t = f1, float(t)

    if best_f1 < 0:
        f1_arr = (2 * prec[:-1] * rec[:-1]) / (prec[:-1] + rec[:-1] + 1e-12)
        i = int(np.argmax(f1_arr))
        best_t = float(thr[i])

    return float(best_t)

def subgroup_metrics(y_true, y_prob, y_pred, races):
    out = []
    races = np.asarray(races).astype(str)
    groups = sorted(set(races.tolist()))
    for g in groups:
        m = (races == g)
        if m.sum() == 0:
            continue
        yt = y_true[m]
        yp = y_prob[m]
        yh = y_pred[m]

        auroc = roc_auc_score(yt, yp) if (yt.min() != yt.max()) else np.nan
        auprc = average_precision_score(yt, yp) if (yt.min() != yt.max()) else np.nan
        p, r, f1, _ = precision_recall_fscore_support(yt, yh, average="binary", zero_division=0)

        out.append({
            "race": g,
            "n": int(m.sum()),
            "pos": int(yt.sum()),
            "auroc": float(auroc) if not np.isnan(auroc) else np.nan,
            "auprc": float(auprc) if not np.isnan(auprc) else np.nan,
            "precision": float(p),
            "recall": float(r),
            "f1": float(f1),
        })
    return out


# -------------------------
# 19) TRAIN LOOP
# -------------------------
def train_model(model, optimizer, name="Model", n_epochs=EPOCHS, alpha_los=ALPHA_LOS):
    model.to(device)
    es = EarlyStoppingAUPRC()

    print(f"\n================ Training {name} on {device} ================")
    hist_loss, hist_auprc = [], []

    for ep in range(1, n_epochs + 1):
        model.train()
        total = 0.0

        for dx_b, dxm_b, stat_b, am_b, ymb, ylb, _ in train_loader:
            dx_b  = dx_b.to(device)
            dxm_b = dxm_b.to(device)
            stat_b = stat_b.to(device)
            am_b  = am_b.to(device)
            ymb   = ymb.to(device)
            ylb   = ylb.to(device)

            optimizer.zero_grad()
            pm, pl = model(dx_b, dxm_b, stat_b, am_b)

            loss_m = bce(pm, ymb)
            loss_l = los_loss_fn(pl, ylb)
            loss = loss_m + alpha_los * loss_l

            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()
            total += loss.item()

        # val AUPRC for early stopping
        _, v_prob, v_y, _, _, _ = collect_logits_probs_and_los(model, val_loader)
        val_auprc = average_precision_score(v_y, v_prob) if (v_y.min() != v_y.max()) else np.nan
        tr_loss = total / max(1, len(train_loader))

        hist_loss.append(tr_loss)
        hist_auprc.append(val_auprc)

        if ep == 1 or ep % 5 == 0:
            print(f"{name} Epoch {ep:02d} | train_loss={tr_loss:.4f} | val_AUPRC={val_auprc:.4f}")

        if es.step(val_auprc):
            print(f"Early stopping at epoch {ep:02d} (best val_AUPRC={es.best:.4f})")
            break

    # Plot training curves (also save)
    ensure_plot_dir()

    plt.figure(figsize=(7, 4))
    plt.plot(hist_loss, label="Train loss")
    plt.title(f"{name} - Train loss")
    plt.xlabel("Epoch"); plt.ylabel("Loss")
    plt.legend(); plt.tight_layout()
    savefig(f"{PLOT_DIR}/{name}_train_loss.png")
    plt.show()

    plt.figure(figsize=(7, 4))
    plt.plot(hist_auprc, label="Val AUPRC")
    plt.title(f"{name} - Val AUPRC")
    plt.xlabel("Epoch"); plt.ylabel("AUPRC")
    plt.legend(); plt.tight_layout()
    savefig(f"{PLOT_DIR}/{name}_val_auprc.png")
    plt.show()

    return model


# -------------------------
# 20) EVALUATION (ROC + Confusion + LOS + subgroup) + SAVE PLOTS
# -------------------------
def evaluate_model(model, name="Model"):
    ensure_plot_dir()

    # pick threshold on VAL using F1
    _, v_prob, v_y, _, _, _ = collect_logits_probs_and_los(model, val_loader)
    thr = pick_threshold_max_f1_min_pos(v_y, v_prob, min_pos=MIN_PRED_POS_VAL)

    # evaluate on TEST
    _, t_prob, t_y, t_lhat, t_ltrue, t_races = collect_logits_probs_and_los(model, test_loader)
    yhat = (t_prob >= thr).astype(int)

    auroc = roc_auc_score(t_y, t_prob) if (t_y.min() != t_y.max()) else np.nan
    auprc = average_precision_score(t_y, t_prob) if (t_y.min() != t_y.max()) else np.nan
    p, rc, f1, _ = precision_recall_fscore_support(t_y, yhat, average="binary", zero_division=0)

    los_mae = float(np.mean(np.abs(t_lhat - t_ltrue)))
    los_rmse = float(np.sqrt(np.mean((t_lhat - t_ltrue) ** 2)))

    print(f"\n=== {name} TEST ===")
    print(f"Chosen VAL thr (max F1): {thr:.4f}")
    print(f"AUROC: {auroc:.4f} | AUPRC: {auprc:.4f} | Prec: {p:.4f} | Rec: {rc:.4f} | F1: {f1:.4f}")
    print("\nClassification report:")
    print(classification_report(t_y, yhat, zero_division=0))
    print(f"{name} LOS MAE (hours):  {los_mae:.2f}")
    print(f"{name} LOS RMSE (hours): {los_rmse:.2f}")

    # Save ROC + Confusion plots
    plot_and_save_roc(
        t_y, t_prob,
        title=f"{name} - ROC Curve (TEST)",
        outpath=f"{PLOT_DIR}/{name}_roc.png",
        label=f"{name} AUROC={auroc:.4f}"
    )
    plot_and_save_confusion(
        t_y, yhat,
        title=f"{name} - Confusion Matrix (TEST)",
        outpath=f"{PLOT_DIR}/{name}_confusion.png"
    )

    # subgroup metrics by race
    sub = subgroup_metrics(t_y, t_prob, yhat, t_races)
    sub_sorted = sorted(sub, key=lambda d: (-d["n"], d["race"]))
    print(f"\n{name} - Subgroup metrics by race (TEST):")
    for row in sub_sorted:
        print(
            f'  race="{row["race"]}" | n={row["n"]} pos={row["pos"]} | '
            f'AUROC={row["auroc"]:.4f} AUPRC={row["auprc"]:.4f} | '
            f'P={row["precision"]:.4f} R={row["recall"]:.4f} F1={row["f1"]:.4f}'
        )

    return {
        "thr": float(thr),
        "auroc": float(auroc) if not np.isnan(auroc) else np.nan,
        "auprc": float(auprc) if not np.isnan(auprc) else np.nan,
        "precision": float(p),
        "recall": float(rc),
        "f1": float(f1),
        "los_mae": float(los_mae),
        "los_rmse": float(los_rmse),
        "subgroups": sub_sorted,
        # return test arrays so we can make combined ROC later
        "test_y": t_y,
        "test_prob": t_prob
    }


# -------------------------
# 21) RUN LSTM + TRANSFORMER
# -------------------------
# LSTM
adm_encoder = AdmissionEncoder(vocab_size=VOCAB_SIZE, dx_emb_dim=64, static_dim=D_static, dropout=0.15)
lstm_core = ClinicalMultiTaskLSTM(input_dim=adm_encoder.out_dim, hidden_dim=128)
model_lstm = LSTMWrapper(adm_encoder, lstm_core)
opt_lstm = optim.AdamW(model_lstm.parameters(), lr=LR_LSTM, weight_decay=WD_LSTM)

model_lstm = train_model(model_lstm, opt_lstm, name="LSTM", n_epochs=EPOCHS, alpha_los=ALPHA_LOS)
lstm_summary = evaluate_model(model_lstm, "LSTM")
print("\nLSTM summary:", {k: v for k, v in lstm_summary.items() if k not in ("subgroups", "test_y", "test_prob")})

# Transformer
if torch.cuda.is_available():
    torch.cuda.empty_cache()

adm_encoder2 = AdmissionEncoder(vocab_size=VOCAB_SIZE, dx_emb_dim=64, static_dim=D_static, dropout=0.15)
trans_core = ClinicalTransformer(input_dim=adm_encoder2.out_dim, embed_dim=128, nhead=4, num_layers=2, dropout=0.25)
model_trans = TransWrapper(adm_encoder2, trans_core)
opt_trans = optim.AdamW(model_trans.parameters(), lr=LR_TRANS, weight_decay=WD_TRANS)

model_trans = train_model(model_trans, opt_trans, name="Transformer", n_epochs=EPOCHS, alpha_los=ALPHA_LOS)
trans_summary = evaluate_model(model_trans, "Transformer")
print("\nTransformer summary:", {k: v for k, v in trans_summary.items() if k not in ("subgroups", "test_y", "test_prob")})

# -------------------------
# 22) COMBINED ROC (LSTM vs Transformer) + SAVE
# -------------------------
ensure_plot_dir()
y_true = lstm_summary["test_y"]  # same ordering in test_loader across runs

fpr1, tpr1, auc1 = roc_points(y_true, lstm_summary["test_prob"])
fpr2, tpr2, auc2 = roc_points(y_true, trans_summary["test_prob"])

plt.figure(figsize=(6, 5))
plt.plot(fpr1, tpr1, label=f"LSTM AUROC={auc1:.4f}")
plt.plot(fpr2, tpr2, label=f"Transformer AUROC={auc2:.4f}")
plt.plot([0, 1], [0, 1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison (TEST)")
plt.legend()
plt.tight_layout()
savefig(f"{PLOT_DIR}/roc_comparison.png")
plt.show()

print("\nSaved plots to:", PLOT_DIR)
print(" - LSTM_train_loss.png, LSTM_val_auprc.png, LSTM_roc.png, LSTM_confusion.png")
print(" - Transformer_train_loss.png, Transformer_val_auprc.png, Transformer_roc.png, Transformer_confusion.png")
print(" - roc_comparison.png")
print("\nDone.")
```
